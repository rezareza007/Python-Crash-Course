## Simple project 1: Computing exp(x)

We know that exp(x)=1+x+x^2/2!+x^3/3!+...
We want to define a function exp(x) that computes e^x


a_n= x^n/(n!)
exp(x)=a_0+a_1+a_2+...

We use the recursion:
a_(n+1)=a_n * (x/(n+1))

```python
def exp(x) : #defining a function
    result=0
    n=0
    a=1 #plays the role of a_n above

    while(a>1e-16):
        result+=a #result=result+a
        #print(a, end='+')
        a=a*(x/(n+1))
        
        n=n+1
        
    return result    

```

```python
print(exp(1))
# Comparing the result of our function to Python's own log function:
math.exp(1)
```

Write a function which takes a list L, and a value s, and removes all the occurence of s in L


```python
def remove_all(L,s): 
    """
    Takes a list L and a value s and removes all the occurances of s from L
    """
    while(s in L) :
        L.remove(s)
    return L    


L=[1,2, 5, 7,5, 2, 6, 9, 4, 1]

remove_all(L, 2)
```

## Simple project 3: Analysing the Diabetes dataset

```python
#The file diabetes.csv contains the information about a sample of women form a native american tribe, such as age, the umber of pregnacies and also whether they have diabetes or not 
# It is available from www.kaggle.com/diabetes/uciml/pima-indians-diabetes-database

f=open("diabetes.csv","r" )
L=f.readlines()
f.close()
L #contains the lines of the file
```

```python
# The names of the columns in the dataset
header=L[0]
header.split()
```

```python
# Turning the data in the file into a 2d list of numbers
data=[]
for line in L[1:]:
    record=line.split(',') #splits each line by commas
    record=[float(x) for x in record]
    print(record)

    data.append(record)
```

```python
data[1] #each element of data corresponds to a line in the file
```

```python
len(data) # number of records in the file
```

```python
#The first column is the number of pregnancies
pregnancies=[x[0] for x in data]
pregnancies
```

```python
#maximum number of pregnancies for one person in the dataset
max(pregnancies) 
```

```python
# We  want to see the relation between the first and the last column
# i.e. between the number of pregnancies and having diabetes or not. 
#For this, we want to compare the average number of pregnancies for the diabeti vs the non-diabetic persons
```

```python
# Computing the mean of a list of numbers

def mean(L):
    sum=0
    for x in L:
        sum=sum+x
    return sum/len(L)    
```

```python
#Average number of pregnancies in the dataset
mean(pregnancies) 
```

```python
#Subset of data containing diabetic people 
# Having diabetes or not, is recorded in the last column
Diabetes=[record for  record in data if record[-1]==1.0 ]

```

```python
#Data of those who don't have diabetes
noDiabetes=[record for  record in data if record[-1]!=1.0 ]
```
